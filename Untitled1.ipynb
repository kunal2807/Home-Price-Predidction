```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams["figure.figsize"] = (20,10)

pd.set_option('display.max_columns', None)
```

```python
df1 = pd.read_csv('./delhi_house_data.csv')
print(df1.shape)

df1 = df1.drop('Unnamed: 0', axis = 'columns')

df1.head()
```

<h1> Handling Null Values </h1>

```python
def print_na_rows(df, subset=''):
    null_mask = df.isna().any(axis=1)
    na_rows = df[null_mask]
    return na_rows
```

```python
print(df1.isna().sum())

print("-----------")
for col in df1:
    print(col, "--> UNIQUE VALUES: ",len(df1[col].unique()))

print("--------\n", df1.shape)
```

```python
print_na_rows(df1).head(3)
```

```python
df2 = df1.drop(['property', 'carpet.area', 'carpet_area_sqft', 'status', 'facing', 'overlooking', 'ownership', 'parking', 'floor'], axis = 'columns')
df2 = df2.dropna(subset=['Rate'])
df2.head()
```

```python
df2.isna().sum()
```

```python
df3 = df2.copy()
df3['balcony'] = df2['balcony'].fillna(df2.bedroom-1)
df3['bathroom'] = df2['bathroom'].fillna(df2.bedroom-1)
df3.isna().sum()
df3.head(3)
```

<h1> Data Cleanup

```python
def isFloat(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
check_float = ['Price', 'bathroom', 'balcony', 'Rate_per_sqft', 'bedroom', 'total_area']

for item in check_float:
    print(df3[~(df3[item].apply(isFloat))])

df4 = df3.drop(['Name', 'Rate'], axis = 'columns')
df4.head(3)
```

```python
df4.location = df4.location.apply(lambda x: x.strip())
df4.head()
```

```python
location_stats = df4.groupby('location')['location'].agg('count').sort_values(ascending = False)
print(len(location_stats))
location_stats.head(30)
```
