```python
def printNullRows(df):
    null_mask = df.isna().any(axis = 1)
    null_rows = df[null_mask]
    print(null_rows)
    
printNullRows(df2)
```

```python
df2[df2.isna()["size"]]
df3 = df2.dropna(subset=["size", "location"])
df3.head()
df3.isna().sum()
```

```python
df3["size_in_bhk"] = df3["size"].apply(lambda x: int(x.split(' ')[0]))
df3["bath"] = df3["bath"].fillna(df3["size_in_bhk"]-1)
df3["balcony"] = df3["balcony"].fillna(df3["size_in_bhk"]-1)
df3.head()
```

```python
df3.isna().sum()
#df3["bath"].unique()
```

```python
df3[ df3.size_in_bhk < df3.bath]
```

```python
def isFloat(x):
    try:
        float(x)
    except:
        return False
    
    return True


df3[~df3.total_sqft.apply(isFloat)].head()
```

```python
def cleanTotalArea(x):

    try:
        tokens = x.split('-')
        if len(tokens)==2:
            return (float(tokens[0]) + float(tokens[1]))/2.0
        
        try:
            return float(x)
        except:
            return None
    
    except:
        return None

    
cleanTotalArea("20")
     
df4 = df3.copy()
df4["total_sqft"] = df4.total_sqft.apply(cleanTotalArea)
df4[~df4.total_sqft.apply(isFloat)].head()
df4.head()
df4.isna().sum()
df4.dropna()
# printNullRows(df4)
df4.count()
```

```python
df4.location = df4.location.apply(lambda x: x.strip())
len(df4.location.unique())
```

```python
location_stats = df4.groupby('location')['location'].agg('count').sort_values(ascending = False)
len(location_stats[location_stats<5])

for x in range(20):
    print(len(location_stats[location_stats == x]), " ", x)

# len(location_stats[location_stats == 5])#.head(100)
location_state_less_10 = location_stats[location_stats <= 10]
```

```python
df4.location = df4.location.apply(lambda x: 'other' if x in location_state_less_10 else x)
df4.head()
len(df4.location.unique())
```

```python
df4.head()
df4['price_per_sqft'] = df4.price / df4.total_sqft
df4.head()
```

```python
print(df4[df4.total_sqft/df4.size_in_bhk < 500].location.count())
print(df4[(df4.bath) > (df4.size_in_bhk + 2)].location.count())
print(df4[(df4.balcony) > (df4.size_in_bhk + 2)].location.count())
print(df4.location.count())

df4[(df4.bath) > (df4.size_in_bhk + 2)].head(30)
```

```python
df5 = df4[~((df4.total_sqft/df4.size_in_bhk < 500) | (df4.bath) > (df4.size_in_bhk + 2))]
df5.head()
```

```python
df5.price_per_sqft.describe()
```

```python
def remove_outliers_pps(df):
    output = pd.DataFrame()
    
    for key, frame in df.groupby('location'):
        mean = np.mean(frame.price_per_sqft)
        stdv = np.std(frame.price_per_sqft)
        
        subFrame = frame[ (frame.price_per_sqft > (mean - stdv)) & (frame.price_per_sqft <= (mean+stdv))]
        
        output = pd.concat([output, subFrame], ignore_index=True)
        
    return output

df6 = remove_outliers_pps(df5)
print(df5.shape)
print(df6.shape)
df6.head()

# df6.location.unique()
```

```python
# loc = 'Yeshwanthpur'


def plot(df, location):
    bhk2 = df[(df.location == location) & (df.size_in_bhk == 2)]
    bhk3 = df[(df.location == location) & (df.size_in_bhk == 3)]
    
    plt.scatter(bhk2.total_sqft, bhk2.price, color='blue', s=50, label = '2 BHK')
    plt.scatter(bhk3.total_sqft, bhk3.price, color='green', s=50, label = '3 BHK', marker = '+')
    
    plt.xlabel('sqft')
    plt.ylabel('Million Rs')
    plt.title(location)
    plt.legend()

    
plot(df6, "Hebbal")
```

```python
#For same location, remove 2 bhk flats whose price per sqft is higher than of 1 bhk
def remove_outliers_bhk(df):
    exclude = np.array([])
    for loc, dfloc in df.groupby('location'):
        bhk_stats = {}
        for bhk, dfbhk in dfloc.groupby('size_in_bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(dfbhk.price_per_sqft),
                'stdv': np.std(dfbhk.price_per_sqft),
                'count': dfbhk.shape[0]
            }
            
            
        for bhk, dfbhk in dfloc.groupby('size_in_bhk'):
            stats = bhk_stats.get(bhk-1)
            
            if stats and stats['count']>5:
                exclude = np.append(exclude, dfbhk[dfbhk.price_per_sqft < stats['mean']].index.values)
                
    return df.drop(exclude, axis = 'index')

df7 = remove_outliers_bhk(df6)
```
